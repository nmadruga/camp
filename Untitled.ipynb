```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
cols = ['TrueA', 'ModelA', 'TrueB', 'ModelB', 'TrueC', 'ModelC', 'TrueH2O', 'ModelH2O']
df = pd.read_csv('/Users/kshores/Downloads/condensed_phase_photolysis_results.txt', 
            delim_whitespace=True, header=None,names=cols)
```

```python
fig, axes = plt.subplots(1,3,figsize=(10,6),dpi=300)

fig.suptitle('model rate == true rate')

(ax1, ax2, ax3) = axes

df[['TrueA', 'ModelA']].plot(ax=ax1, alpha=0.5)
df[['TrueB', 'ModelB']].plot(ax=ax2, alpha=0.5)
df[['TrueC', 'ModelC']].plot(ax=ax3, alpha=0.5)

for ax in axes:
    ax.spines[:].set_visible(False)
    ax.grid()
    ax.tick_params(width=0)
```
